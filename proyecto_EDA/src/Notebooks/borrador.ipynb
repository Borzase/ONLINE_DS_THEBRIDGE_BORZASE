## En primer lugar cargamos librerias, importamos el fichero y limpiamos los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel("datos_vibra.xlsx", usecols="D:O", skiprows=[0,1,2])
df
```

```python
df.loc[df["Diagnostico 1"].isna(), "Diagnostico 1"] = "Acondicionamiento"
df
```

```python
df.loc[df["Diagnostico 1"] == "Acondicionamiento Deportivo", "Diagnostico 1"] = "Acondicionamiento"
df
```

```python
df["Diagnostico 1"].value_counts()
```

```python
condicion_1 = df["Diagnostico 1"] == "Mielopatia"
condicion_2 = df["Diagnostico 1"] == "Mielopatia Degenerativa"
df.loc[condicion_1 | condicion_2, "Diagnostico 1"]= "Mielopatia"
df["Diagnostico 1"].value_counts()
```

```python
df.loc[df["Diagnostico 1"] == "Mielopatia ", "Diagnostico 1"] = "Mielopatia"
df["Diagnostico 1"].value_counts()
```

```python
df.loc[df["Diagnostico 1"] == "Sobrepeso", "Diagnostico 1"] = "Acondicionamiento"
df["Diagnostico 1"].value_counts()
```

```python
df.loc[df["Diagnostico 1"] == "Necrosis Avascular Cabezafemoral", "Diagnostico 1"] = "NAC"
df["Diagnostico 1"].value_counts()
```

```python
df.info()
```

```python
df.loc[df["Gasto_rehabilitacion"].isna(), "Gasto_rehabilitacion"] = 70
df.info()
```

```python
df.loc[df["Gasto_tienda"].isna(), "Gasto_tienda"] = 0
df.info()
```

```python
df.loc[df["Dieta"].isna(), "Dieta"] = 0
df.drop(columns=["Diagnóstico 2"], inplace = True)
df.info()
```

```python
print(df.Raza.nunique())
print(df.Raza.value_counts())
```

```python
orden = df["Diagnostico 1"].value_counts().index
valores = df["Diagnostico 1"].value_counts(normalize = True) * 100
fig,axs = plt.subplots(nrows=2,ncols=1, figsize=(15,10))
fig.suptitle("Frecuencias Diagnóstico")
sns.countplot(x = "Diagnostico 1", data = df, ax = axs[0], hue = "Diagnostico 1",palette="husl", legend = False, order=df["Diagnostico 1"].value_counts().index)            # hue es necesario para diferenciar con color cada barra. Si no se pone, colorea todas las barras iguales. Por eso hue se iguala a la variable que hay en x
axs[0].set_title("Frecuencias Absolutas")
sns.barplot(x = valores.index, y = valores.values, hue = valores.index, ax = axs[1], palette="husl")
axs[1].set_title("Frecuencias relativas")
axs[1].set_ylabel("%")
for i in range(len(valores)):
    axs[1].text(df["Diagnostico 1"].value_counts().index[i], valores[i], f"{valores[i].round(2)}%", ha="center")

plt.tight_layout();
```

```python
diccion = {}
for valor in df_prueba["Raza"].unique():
    diccion[valor] = df_prueba.loc[df_prueba["Raza"] == valor, "Diagnostico 1"]
```

```python
diccion
```

```python
fig, axs = plt.subplots(nrows = 5, ncols = 1, figsize = (10,20))
n = 0
for clave, valor in diccion.items():
    
    valores = valor.value_counts(normalize = True) * 100
    sns.barplot(x = valores.index, y = valores.values, hue = valores.index, palette="husl", ax=axs[n])
    axs[n].set_title(f"Frecuencias relativas de {clave}")
    axs[n].set_ylabel("%")
    for i in range(len(valores)):
        axs[n].text(valor.value_counts().index[i], valores[i], f"{valores[i].round(2)}%", ha="center")
    n += 1
plt.tight_layout();
```
