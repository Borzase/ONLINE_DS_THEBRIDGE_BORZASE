```python
def chaqueta():
    chaqueta.llamar_web()
    chaqueta.relleno_tabla("chaqueta")
    chaqueta.solicitud()

    filtrado = input("¿Desea filtrar por modelo, color o precio? Si o No")
    while filtrado == "Si":
        opcion = int(input("""¿Cómo desea filtrar?:
                       1 - Modelo
                       2 - Color
                       3 - Precio
                       4 - Salir """))
        if opcion == 1:
            a = input("Escriba el modelo de la prenda.")
            chaqueta.solicitud(modelo = a)
            continue
        elif opcion == 2:
            b = input("Escriba el color de la prenda.")
            chaqueta.solicitud(color = b)
            continue
        elif opcion == 3:
            c = input("Escriba el precio referente de la prenda. Se seleccionaran prendas por valor igual o menor al precio escogido")
            chaqueta.solicitud(precio = c)
            continue
        elif opcion == 4:
            break
        break


          
```
