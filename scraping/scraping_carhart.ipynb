```python
import pandas as pd
import requests

from bs4 import BeautifulSoup
import sqlite3
```

```python
url = "https://www.carhartt-wip.com/es/hombre-seasonal-bestseller?country=es&trc_gcmp_id=10948255172&trc_gag_id=107054753266&trc_gad_id=645062614224&gad_source=1&gclid=CjwKCAjw5PK_BhBBEiwAL7GTPYzx3IPuzz_JUFjNVVHoFlZp4aaSz42NzxXVRqWCqvDAhhPPVoCUbRoCzdwQAvD_BwE"
```

```python
real = requests.get(url) # Obtenemos la página dentro de un objeto Request

soup = BeautifulSoup(real.text, "lxml")
```

```python
elementos = soup.find_all("article")
len(elementos)
```

```python
datos = []
```

```python
for elem in elementos:
    modelo = elem.find("p", attrs={"class":"descr"})
    color = elem.find("p", attrs={"class":"color"})
    precio = elem.find("span", attrs={"class": "country_eur"})
    datos.append({
        "Modelo": modelo.text,
        "Color": color.text,
        "Precio": precio.text
    })
```

```python
df = pd.DataFrame(datos)
```

```python
df
```

```python
df["Category"] = "Accesories"

df.loc[df["Modelo"].str.contains("Jacket"), "Category"] = "Jacket"
df.loc[df["Modelo"].str.contains("Socks"), "Category"] = "Socks"
df.loc[df["Modelo"].str.contains("Short"), "Category"] = "Short"
df.loc[df["Modelo"].str.contains("Sweatshirt"), "Category"] = "Sweatshirt"
df.loc[df["Modelo"].str.contains("Pant"), "Category"] = "Pant"
df.loc[df["Modelo"].str.contains("Shirt"), "Category"] = "Shirt"
df.loc[df["Modelo"].str.contains("T-Shirt"), "Category"] = "T-Shirt"
```

```python
conexion = sqlite3.connect("base_datos_chaqueta.db")                        # Creo una tabla sqlite para poder hacer consultas por pantalla
cursor = conexion.cursor()
query_create = ''' 
CREATE TABLE Tabla (
MODELO     TEXT NOT NULL,
COLOR      TEXT NOT NULL,
PRECIO      FLOAT,


) '''
cursor.execute(query_create)



```

```python
for unidad in datos:
    Modelo = unidad["Modelo"]
    Color = unidad["Color"]
    Precio = unidad["Precio"]
    query = f"INSERT INTO Tabla (Modelo, Color, Precio) VALUES ('{Modelo}', '{Color}', '{Precio}')"
    cursor.execute(query)
```

```python
conexion.commit()
```
