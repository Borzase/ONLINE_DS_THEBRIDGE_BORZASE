### #1.1

Carga el dataset de casas de California que está en la ruta "./data/california_cities.csv". Muestra parte de su contenido y su descripción. Deshazte de las filas con nulos.

```python
cities = pd.read_csv("./data/california_cities.csv")
cities
```

```python
df_cities = cities.copy()
df_cities.dropna(axis = "index", inplace = True)
df_cities.info()
```

```python
df_cities.population_total.sort_values(ascending = False)
```

```python
ciudad_1= df_cities["Unnamed: 0"] == 239
ciudad_2 = df_cities["Unnamed: 0"] == 367
ciudad_3 = df_cities["Unnamed: 0"] == 359
ciudad_4 = df_cities["Unnamed: 0"] == 9
ciudad_5 = df_cities["Unnamed: 0"] == 386
condicion = ciudad_1 | ciudad_2 | ciudad_3 | ciudad_4 | ciudad_5
mis_ciudad = df_cities.loc[condicion]
mis_ciudad.sort_values("population_total", ascending= False, inplace= True)
mis_ciudad
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
ciudades = mis_ciudad.city.to_list()
valores = mis_ciudad.population_total.to_list()
fig = plt.figure()
plt.bar(ciudades, valores, color = ["mediumslateblue","grey", "olive", "brown", "salmon"], alpha = 0.8)
plt.title("Ciudades más pobladas de California")
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.ylim(0, 5000000)

for i in range(len(ciudades)):
    plt.text(ciudades[i], valores[i], f"{valores[i]}", ha="center")
```

```python
plt.hist(df_cities["area_total_km2"], bins = 50, color="steelblue", label = "Hist Area")
plt.legend()
font_dict={
    "size" : 18,
    "family": "Bradley Hand itc"
}
plt.xlabel("Superficie ($km^2$)", fontdict = font_dict)
plt.ylabel("Numero de ciudades", fontdict = font_dict);
```

```python
plt.scatter(df_cities["area_water_km2"],  df_cities["population_total"], c ="g", alpha=0.5)
plt.xlabel("Area de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacion (miles)")
plt.title("Poblacion por kilómetro cuadrado de agua")
plt.annotate("Outliers", xy =(88, 3884307), xytext=(60, 3000000), arrowprops=dict(arrowstyle = "->", connectionstyle = "arc3"))
plt.annotate("Outliers", xy =(122.270, 1345895), xytext=(60, 3000000), arrowprops=dict(arrowstyle = "->", connectionstyle = "arc3"))
```

```python
df_cities.area_water_km2.sort_values(ascending=False)
```

```python

plt.boxplot(df_cities.elevation_m, vert=False, patch_artist= True, boxprops={"facecolor": "red"}, medianprops={"color":"black"})
plt.xlabel("Altura(m)")
plt.title("Diagrama de caja de la altura de las ciudades de California")
```

```python
fig,ax = plt.subplots(2,2, figsize = (15,12))
fig.subplots_adjust(hspace= 0.4, wspace = 0.2)
ax[0,0].bar(ciudades, valores, color = ["mediumslateblue","grey", "olive", "brown", "salmon"], alpha = 0.8)
ax[0,0].set_title("Ciudades más pobladas de California")
ax[0,0].set_xlabel("Ciudades")
ax[0,0].set_ylabel("Habitantes (miles)")
for i in range(len(ciudades)):
    ax[0,0].text(ciudades[i], valores[i], f"{valores[i]}", ha="center")

ax[0,1].hist(df_cities["area_total_km2"], bins = 50, color="steelblue", label = "Hist Area")
ax[0,1].legend()
font_dict={
    "size" : 18,
    "family": "Bradley Hand itc"}
ax[0,1].set_xlabel("Superficie ($km^2$)", fontdict = font_dict)
ax[0,1].set_ylabel("Numero de ciudades", fontdict = font_dict)

ax[1,0].scatter(df_cities["area_water_km2"],  df_cities["population_total"], c ="g", alpha=0.5)
ax[1,0].set_xlabel("Area de agua en la ciudad ($km^2$)")
ax[1,0].set_ylabel("Poblacion (miles)")
ax[1,0].set_title("Poblacion por kilómetro cuadrado de agua")
ax[1,0].annotate("Outliers", xy =(88, 3884307), xytext=(60, 3000000), arrowprops=dict(arrowstyle = "->", connectionstyle = "arc3"))
ax[1,0].annotate("Outliers", xy =(122.270, 1345895), xytext=(60, 3000000), arrowprops=dict(arrowstyle = "->", connectionstyle = "arc3"))

ax[1,1].boxplot(df_cities.elevation_m, vert=False, patch_artist= True, boxprops={"facecolor": "red"}, medianprops={"color":"black"})
ax[1,1].set_xlabel("Altura(m)")
ax[1,1].set_title("Diagrama de caja de la altura de las ciudades de California")
```
