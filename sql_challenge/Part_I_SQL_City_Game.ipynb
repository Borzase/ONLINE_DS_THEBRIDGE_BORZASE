A crime has taken place and the detective needs your help. The detective gave you the crime scene report, but you somehow lost it. You vaguely remember that the crime was a murder that occurred sometime on Jan.15, 2018 and that it took place in SQL City. Start by retrieving the corresponding crime scene report from the police department’s database.

```python
import pandas as pd
import sqlite3
```

```python
connection = sqlite3.connect("./data/sql-murder-mystery.db")
cursor_comisario = connection.cursor()
```

```python
def sql_query(query):
    cursor_comisario.execute(query) 
    ans = cursor_comisario.fetchall()
    names = [d[0] for d in cursor_comisario.description]
    df_murder = pd.DataFrame(ans,columns=names)
    return df_murder 
```

```python
query= '''  
SELECT *
FROM sqlite_master'''

sql_query(query)
```

```python
query = '''  
SELECT *
FROM crime_scene_report
'''

sql_query(query)
```

```python
query = '''  
SELECT *
FROM crime_scene_report
WHERE date IN (20180115) AND city LIKE "%SQL City%" AND type LIKE "murder"
'''

df = sql_query(query)
df.description.to_list()
```

```python
query = '''  
SELECT name, address_street_name, address_number
FROM person
WHERE address_street_name IN ("Northwestern Dr") OR (address_street_name IN ("Franklin Ave") AND name LIKE "%Annabel%")
ORDER BY 3
'''

sql_query(query)
```

```python
query = '''  
SELECT *
FROM person AS a
LEFT JOIN interview AS b
ON a.id = b.person_id
WHERE a.name IN ("Annabel Miller", "Morty Schapiro")
'''

sql_query(query)
```

```python
df_entrevista = sql_query(query)
df_entrevista.transcript.to_list()
```

```python
query = '''  
SELECT a.id, a.name, a.membership_status, a.person_id, b.membership_id, b.check_in_date
FROM get_fit_now_member AS a
LEFT JOIN get_fit_now_check_in AS b
ON a.id = b.membership_id
WHERE a.membership_status LIKE "gold" AND a.id LIKE "48Z%" AND b.check_in_date LIKE "20180109"
'''

sql_query(query)
```

```python
query = '''   
SELECT a.id, a.name, a.license_id, b.id, b.plate_number
FROM person AS a
LEFT JOIN drivers_license AS b
ON a.license_id = b.id
WHERE a.name IN ("Joe Germuska", "Jeremy Bowers") --AND b.plate_number LIKE "%H42W%"
'''

sql_query(query)
```
